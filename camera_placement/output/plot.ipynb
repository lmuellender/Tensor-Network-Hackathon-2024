```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

```

```python
file=open('tn_mps_dmrg_unc.dat')
lines = file.readlines()
file.close()

data = np.zeros((len(lines)-1,5))
sol = []
cameras = []
for line in range(len(lines)-1):
    data[line,:] = np.array(lines[line+1].split(' ')[:5],dtype=float)
    sol.append(np.array([i for i in (''.join(lines[line+1].split(' ')[5:])[3:-2:4])],dtype=int))
cameras = ([len(s) - np.sum(s) for s in sol])
```

```python
file.close()
```

```python
mpl.rcParams["figure.dpi"] = 100
plt.rcParams["figure.figsize"] = (6,4)
plt.plot(data[:,0], data[:,3], 'x', color="purple",label='TTN_DMRG')
# plt.plot(sizes, ed_time, '+', color="green",label='ED')
plt.legend()
plt.xlabel("$N$")
plt.ylabel("$ Runtime $")
# plt.savefig('runtime.png',bbox_inches='tight')
# plt.close()
```
